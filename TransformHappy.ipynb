```python
#Dependencies

import pandas as pd
from sqlalchemy import create_engine
```

```python
#Read Happiness Index csv

csv_file = "resources/2019.csv"
happy_2019_df = pd.read_csv(csv_file)
happy_2019_df.head()
```

```python
#Read Best Cities csv

file = "resources/Movinga_best_cities.csv"
cities2019_df = pd.read_csv(file)
cities2019_df.head()
```

```python
#Get rid of irrelevant columns in cities2019

new_cities = cities2019_df.drop(columns=['Total', 'Lat', 'Long'])
new_cities.head()
```

```python
#Connect to database for cities db

rds_connection_string = "postgres:Password@localhost:5432/Happy_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Load cities csv to database

new_cities.to_sql(name='cities', con=engine, if_exists='append', index=False)
```

```python
#Load happiness csv to database

happy_2019_df.to_sql(name='happiness', con=engine, if_exists='append', index=False)
```

```python
#Check for tables

pd.read_sql_query('select * from cities', con=engine).head()
```
